Note that what exception types exist depends on what modules have been imported, and thus on how this notebook is run.

```python
def get_descendants(root: type) -> set[type]:
    descendants = set[type]()

    def dfs(parent: type) -> None:
        if parent in descendants:
            return
        descendants.add(parent)
        for child in parent.__subclasses__():
            dfs(child)

    dfs(root)
    return descendants
```

## All exceptions

```python
get_descendants(BaseException)
```

## Exceptions *not* inheriting from `Exception`

```python
get_descendants(BaseException) - get_descendants(Exception)
```
